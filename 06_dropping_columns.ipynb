# Dropping Columns
#### Pandas drop()


```python
import pandas as pd

data = pd.read_csv('./data/artwork_sample.csv')
```

```python
data.head()
```

```python
# dropping rows - inplace is False by default
data.drop(0, inplace=False)
```

```python
data
```

```python
# KeyError: "['id'] not found in axis"
data.drop('id')
```

```python
data.drop('id', axis=1)
```

```python
# columns can be dropped with columns param instead of axis as well
data.drop(columns=['id'])
```

```python
data.drop(columns=['id', 'height', 'width', 'depth'])
```

```python
# removing rows with labels parameter
data.drop(labels=[0, 1, 2])
```

```python
help(data.drop)
```

```python
import numpy as np
data.drop(labels=np.arange(2,6))
```

```python
data.drop(columns=['id'], inplace=True)
```

```python
data
```

```python
# since data does not have id column now, if we try to drop it again, it will raise error
data.drop(columns=['id'])
```

```python
# read only required columns of data from csv file
data = pd.read_csv('./data/artwork_sample.csv', usecols=['artist', 'title'])
data
```
